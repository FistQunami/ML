### 1.2 Сравнение моделей (4 балла)

Задача классификации. Все признаки уже числовые. Значит можно пробовать просто все модели и выбрать лучшую. Так и поступим, сделайте кросс валидацию на 5 фолдах, используя `sklearn.model_selection.KFold` как аргумент у `cross_val_score`. Метрика качества будет `accuracy`.

Алгоритмы для тестирования:
- KNeighborsClassifier с 10 соседями
- KNeighborsClassifier с 10 соседями и масштабированием StandartScaler
- RidgeClassifier
- DecisionTreeClassifier 
- BaggingClassifier c 100 деревьев
- BaggingClassifier с 100 деревьев и каждое дерево обучается только по половине случайно выбранных признаков (см аргументы)
- RandomForestClassifier c 100 деревьев

Выведите среднее значение метрики качества для каждого из классификаторов. 

**hint**: каждый следующий алгоритм, будет показывать качество лучше, чем предыдущий. Если у вас не так - то что-то вы делаете неправильно. Везде зафиксируйте random_state=42.

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.pipeline import Pipeline
import numpy as np

# Инициализация KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Список классификаторов
classifiers = {
    "KNeighborsClassifier (k=10)": KNeighborsClassifier(n_neighbors=10),
    
    "KNeighborsClassifier (k=10) + StandartScaler": Pipeline([
        ("scaler", StandardScaler()),
        ("knn", KNeighborsClassifier(n_neighbors=10))
    ]),
    
    "RidgeClassifier": RidgeClassifier(random_state=42),
    
    "DecisionTreeClassifier": DecisionTreeClassifier(random_state=42),
    
    "BaggingClassifier (100 trees)": BaggingClassifier(
        estimator=DecisionTreeClassifier(random_state=42), 
        n_estimators=100, random_state=42
    ),
    
    "BaggingClassifier (100 trees, 50% features)": BaggingClassifier(
        estimator=DecisionTreeClassifier(random_state=42), 
        n_estimators=100, max_features=0.5, random_state=42
    ),
    
    "RandomForestClassifier (100 trees)": RandomForestClassifier(
        n_estimators=100, random_state=42
    )
}

# Оценка классификаторов
results = {}
for name, model in classifiers.items():
    scores = cross_val_score(model, X, y, cv=kf, scoring="accuracy")
    results[name] = np.mean(scores)
    print(f"{name}: {np.mean(scores):.4f}")

print("\nСредние значения accuracy для каждого классификатора:")
for name, score in results.items():
    print(f"{name}: {score:.4f}")
```

### 1.3 Расуждения (8 баллов)

Ответьте на вопросы развернуто, можете полистать литературу:

- почему наблюдается значимая разница в качестве у KNeighborsClassifier с масштабированием и без
- почему масштабирование не важно для деревьев решений
- почему бэггинг на половине признаков для каждого дерева дал качество предсказания больше, чем на всех? (а он дал!)
- у какой модели наибольшей отклонение от среднего качества предсказаний? А почему??

```python
#Почему наблюдается значимая разница в качестве у KNeighborsClassifier с масштабированием и без?
#Алгоритм KNeighborsClassifier основан на измерении расстояния между точками (например, Евклидово расстояние). Если признаки имеют разный масштаб, например, 
#один в диапазоне от 0 до 1, а другой от 0 до 1000, то масштабные признаки будут доминировать при подсчёте расстояния. 
#Это приводит к неправильным результатам классификации, поскольку "дальний" масштаб влияет гораздо больше.
#Масштабирование (например, StandardScaler) приводит все признаки к одному диапазону со средним 0 и стандартным отклонением 1, 
#делая их равнозначными при вычислении расстояний. В результате модель KNN может корректно учитывать влияние всех признаков, что и объясняет улучшение качества.

#Почему масштабирование не важно для деревьев решений?
#Деревья решений, такие как DecisionTreeClassifier, работают путём попыток разделить данные на основе пороговых значений признаков. 
#Они не зависят от конкретных расстояний или масштабов признаков, так как ключевым является позиция значения относительно других значений, а не его абсолютный размер.
#Проще говоря, независимо от того, выражен признак в долях или в тысячах, дерево будет находить оптимальное пороговое значение, делящее данные на классы. 
#Поэтому масштабирование не влияет на производительность деревьев решений.

#Почему бэггинг на половине признаков для каждого дерева дал качество предсказания больше, чем на всех? (а он дал!)
#Когда в бэггинге каждое дерево обучается на случайной подвыборке признаков, это способствует разнообразию деревьев. 
#Разные комбинации признаков приводят к построению различных деревьев, которые могут уловить разные аспекты данных и компенсировать ошибки друг друга при усреднении.
#Если использовать все признаки для каждого дерева, все деревья становятся более похожими друг на друга (коррелированными), что уменьшает эффект ансамблирования. 
#Подвыборка признаков (например, 50%) снижает корреляцию между деревьями, улучшая итоговое качество предсказания благодаря большему разнообразию модели.

#У какой модели наибольшей отклонение от среднего качества предсказаний? А почему??
#Наибольшее отклонение от среднего качества предсказаний обычно наблюдается у DecisionTreeClassifier. 
#Это связано с тем, что одиночные деревья решений сильно подвержены переобучению, особенно на малых выборках данных.
#Деревья решений легко подстраиваются под шум и особенности обучающего набора, что приводит к нестабильности их предсказаний на разных фолдах кросс-валидации. 
#У ансамблевых методов, таких как Bagging или Random Forest, эта проблема решается за счёт усреднения предсказаний, что снижает разброс.
```

### 2.2 Перебор классификаторов (3 балла)

В этом задании вам снова придется перебрать несколько классификаторов, но теперь мы обратим внимание на другие гиперпараметры и их влияние на качество классификации, кстати опять `accuracy`.

Сделайте кроссвалидацию на 10 фолдах, указав `cv=10` для следующих классификаторов:

- DecisionTreeClassifier с параметрами по-умолчанию
- BaggingClassifier с 100 деревьвев
- BaggingClassifier с 100 деревьев, НО с ограничением на максимальное количество признаков, участвующих при обучении каждого из деревьев в $\sqrt{N}$, где $N$ - это число признаков.
- BaggingClassifier с 100 деревьев, НО с ограничением на количество признаков участвующих в разбиении для каждого из деревьев в $\sqrt{N}$, где $N$ - это число признаков. Это отличается от предыдущей модели тем, где ограничивается `max_features`. Читайте документацию :trollface:
- обычный случайный лес со 100 деревьями

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
import numpy as np

digits = load_digits()
X, y = digits.data, digits.target

max_features_sqrt = int(np.sqrt(X.shape[1]))

classifiers = {
    "DecisionTreeClassifier": DecisionTreeClassifier(random_state=42),
    
    "BaggingClassifier (100 trees)": BaggingClassifier(
        estimator=DecisionTreeClassifier(random_state=42),
        n_estimators=100, random_state=42
    ),
    
    "BaggingClassifier (100 trees, max_features=sqrt(N))": BaggingClassifier(
        estimator=DecisionTreeClassifier(random_state=42),
        n_estimators=100, max_features=max_features_sqrt, random_state=42
    ),
    
    "BaggingClassifier (100 trees, max_features=sqrt(N) for split)": BaggingClassifier(
        estimator=DecisionTreeClassifier(max_features=max_features_sqrt, random_state=42),
        n_estimators=100, random_state=42
    ),
    
    "RandomForestClassifier (100 trees)": RandomForestClassifier(
        n_estimators=100, random_state=42
    ),
}

results = {}
for name, clf in classifiers.items():
    score = cross_val_score(clf, X, y, cv=10, scoring='accuracy') 
    results[name] = np.mean(score)  

for name, score in results.items():
    print(f"{name}: {score:.4f}")
```

### 2.4 Количество деревьев (2 балла)

Сделайте перебор количества деревьев для `RandomForestClassifier`. Сохраните качества кросс валидации на 10 фолдах для `[1,5,10,15,50,100,150,200,300]` количества деревьев. Нарисуйте график, где по оси x - количество деревьев, а по оси y - качество. При каком количестве деревьев получается самое хорошее качество?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits

digits = load_digits()
X, y = digits.data, digits.target

n_estimators_values = [1, 5, 10, 15, 50, 100, 150, 200, 300]
cv = KFold(n_splits=10, shuffle=True, random_state=42)

scores = []

for n_estimators in n_estimators_values:
    rf = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    cv_score = cross_val_score(rf, X, y, cv=cv, scoring='accuracy')
    scores.append(np.mean(cv_score))

plt.figure(figsize=(8, 6))
plt.plot(n_estimators_values, scores, marker='o', linestyle='-', color='b')
plt.xlabel('Количество деревьев', fontsize=12)
plt.ylabel('Качество (Accuracy)', fontsize=12)
plt.title('Перебор количества деревьев для RandomForestClassifier', fontsize=14)
plt.grid(True)
plt.show()

best_n_estimators = n_estimators_values[np.argmax(scores)]
best_score = max(scores)
print(f"Лучшее качество достигнуто при количестве деревьев: {best_n_estimators} (качество = {best_score:.4f})")
```

### 2.5 Количество признаков  (2 балла)

Переберите теперь максимальное количество признаков для `RandomForestClassifier` на 100 деревьях, от 1 до 64 с шагом 5. Постройте график качества по кроссвалидации на 10 фолдах.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
digits = load_digits()
X, y = digits.data, digits.target

feature_counts = range(1, 65, 5)

accuracies = []

for n_features in feature_counts:
    model = RandomForestClassifier(n_estimators=100, max_features=n_features, random_state=42)
    scores = cross_val_score(model, X, y, cv=10, scoring="accuracy")
    accuracies.append(np.mean(scores))

plt.plot(feature_counts, accuracies)
plt.xlabel("Количество признаков (max_features)")
plt.ylabel("Качество предсказания")
plt.title("Зависимость качества от числа признаков")
plt.show()
```

### 2.5 Вопросы по RandomForest (8 баллов)

Ответьте на вопросы:

- Что происходит с ростом числа деревьев у случайного леса. Можно ли просто всегда брать 5000 деревьев и быть счастливым?
- Как зависит качество предсказания в дереве в зависимости от max_features?
- Почему качество зависит от max_features?
- Как глубина деревьев влияет на качество случайного леса?

Что происходит с ростом числа деревьев у случайного леса. Можно ли просто всегда брать 5000 деревьев и быть счастливым?
С ростом числа деревьев качество предсказания обычно улучшаетcя, потому что модель становится менее чувствительной к случайным вариациям. Однако, после определённого числа деревьев, прирост в качестве становится минимальным, а время вычислений и использование памяти увеличиваются. Поэтому брать слишком много деревьев может быть неоправданно.

Как зависит качество предсказания в дереве в зависимости от max_features?
При увеличении max_features (количества признаков, рассматриваемых при разбиении дерева) дерево может лучше изучать зависимость между признаками, что обычно улучшает качество. Однако, если значение слишком велико, может происходить переобучение, потому что модель слишком сильно подстраивается под особенности данных.

Почему качество зависит от max_features?
Качество зависит от max_features, потому что параметр влияет на разнообразие деревьев в случайном лесе. Чем меньше признаков выбирается для обучения каждого дерева, тем больше различий между деревьями, что улучшает обобщающую способность случайного леса. Но если количество признаков слишком низкое, дерево не будет достаточно информативным.

Как глубина деревьев влияет на качество случайного леса?
Глубина дерева напрямую влияет на способность модели обучать сложные зависимости. Меньшая глубина может привести к недообучению (underfitting), а большая — к переобучению (overfitting). Оптимальная глубина помогает модели извлекать достаточную информацию из данных, не адаптируясь слишком сильно к шуму.
